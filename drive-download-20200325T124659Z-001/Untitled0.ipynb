```
from pathlib import Path
import pandas as pd
import os
from IPython.display import display


dataset_path = Path("/content/gdrive/My Drive/dataset_short")
```

```
annot_path = dataset_path/"annotations/train/2018-02-13_1418_left/"

lst = os.listdir(annot_path)
lst.sort()

idx = 15

lab = lst[idx]
print(lab)

label_df = pd.read_csv(annot_path/lab, sep='\t')
label_df
```

```
import PIL
import cv2
import numpy as np
from google.colab.patches import cv2_imshow


img_path = dataset_path/"images/train/2018-02-13_1418_left/"
lst = os.listdir(img_path)
lst.sort()
img = PIL.Image.open(img_path/lst[idx])
img = np.float32(img)

# cv2.resize(img, )
for i in range(len(label_df)):
    start_point = (label_df['xtl'].values[i], label_df['ytl'].values[i])
    end_point = (label_df['xbr'].values[i], label_df['ybr'].values[i])
    cv2.rectangle(img, start_point, end_point, 255, 4)

cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(img)
```
